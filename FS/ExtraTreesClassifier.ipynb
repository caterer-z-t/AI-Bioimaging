```python
# Imports
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.datasets import make_classification

import numpy as np
import pandas as pd

import csv 
```

```python
# Filepath
dataFolder = 'Data/'
dataFile = dataFolder + 'bandsOnTopWithClassification.csv'

# Load and fit the dataset
df = pd.read_csv(dataFile)
indColumn = df.iloc[:,0:767] #independent columns
targetColumn = df.iloc[:,-1] #target column
```

```python
# Run the model one time

for i in range(1):
    # Fit the dataset
    model = ExtraTreesClassifier(random_state = i)
    model.fit(indColumn,targetColumn)
    
    # NOTE model.predict may be of use, although we get the predicted class probabilities currently

    output = model.feature_importances_
    sort_index = np.argsort(output)

    saveOutput = 'ExtraTreesOutput_Seed_' + str(i) + '.csv'
    
    with open(saveOutput, 'w') as csvfile:  
        # creating a csv writer object  
        csvwriter = csv.writer(csvfile)  
          
        # writing the fields  
        csvwriter.writerow(model.feature_importances_)  
          
        # writing the data rows  
        #csvwriter.writerows(indColumn)
```
